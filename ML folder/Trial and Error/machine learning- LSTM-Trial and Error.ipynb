```python
#Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as gu
import numpy as np
from datetime import datetime as dt
import plotly.express as px
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
companies = ["AAPL","AMZN","JPM","ADBE","T","IBM","GS","AMD","BLK","HCA","MO", 
             "GD","MU","EQIX","F","HSY","EA","ABC","GPN","AAL"]
            
```

```python
#importing stock data 
stock_data_df = pd.read_csv('modified_stock_data.csv')
stock_data_df.head(10)
```

```python
stock_data_df["Symbol"]
```

```python
stock_data_df.dropna()
```

```python
stock_data_df["Date"] = pd.to_datetime(stock_data_df["Date"], format="%Y-%m-%d")
stock_data_df["Date"].astype
```

```python
stock_data_df["Symbol"].sort_values(ascending=True)
```

```python
stock_data_df = stock_data_df.sort_values(by=["Date","Symbol"], ascending=True)
stock_data_df.reset_index(inplace=True, drop=True)
stock_data_df
```

```python
filt = (stock_data_df["Symbol"] == 0)
apple_df = stock_data_df[filt]
apple_df.reset_index(inplace = True, drop=True)
apple_close = apple_df[["Close"]]
apple_close
```

```python
apple_df["Outcome"]=apple_df.rolling(2).apply(lambda x: x.iloc[1] > x.iloc[0])["Close"]
apple_df.dropna(inplace=True)
apple_df
```

```python
#Define the features
X = apple_df.copy()
X = X.drop(columns = ["Outcome","Date", "Symbol"], axis=1)
#Define the target
y = apple_df["Outcome"].ravel()
y[:5]
```

```python
#Splitting into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=5)
```

```python
#Create standard scaler
#apple_scaled = pd.DataFrame()
#scaler=StandardScaler()
#apple_scaled[["Close", "High", "Low", "Open", "Volume"]] = scaler.fit(apple_df[["Close", "High", "Low", "Open", "Volume"]])
#X_train_scaled = apple_scaled.transform(X_train)
#X_test_scaled = apple_scaled.transform(X_test)

scaler = StandardScaler()
#Fitting scaler with data
X_scaler = scaler.fit(X_train)

#Scaling the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
#Create a random forest classifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=5)
```

```python
# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_train)
```

```python
#Making predictions using testing data
predictions = rf_model.predict(X_test_scaled)
predictions
```

```python
#Calculating the confusion matrix
cm = confusion_matrix(y_test, predictions)

cm_df = pd.DataFrame(cm, index=["Actual 0(decrease)", "Actual 1(increase)"], columns = ["Predicted 0(decrease)", "Predicted 1(increase)"])

cm_df
```

```python
acc_score = accuracy_score(y_test, predictions)
acc_score
```

```python
print(f"{companies[0]}:{classification_report(y_test, predictions)}")
```

```python
from sklearn.ensemble import GradientBoostingClassifier
learning_rates = [0.5]
acc_dict=[]
boost_dict=[]
for i in range(20):
    #Filter original df by symbol # and place results into list_ variable
    filt = (stock_data_df["Symbol"] == i)
    list_ = stock_data_df[filt]
    list_.reset_index(inplace = True, drop=True)
    #Create dataframe from close column of list_ variable
    list_close = list_[["Close"]]
    #print(f"{companies[i]}: {list_close}")
    
    #Create new column called Outcome where 1=price increase and 0=price decrease
    list_["Outcome"]=list_.rolling(2).apply(lambda x: x.iloc[1] > x.iloc[0])["Close"]
    list_.dropna(inplace=True)
    #print(list_)
    
    #Define the features
    X = list_.copy()
    X = X.drop(columns = ["Outcome","Date", "Symbol"], axis=1)
    #Define the target
    y = list_["Outcome"].ravel()
    #print(y[:5])
    
    #Splitting into train and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=5)
    
    scaler = StandardScaler()
    #Fitting scaler with data
    X_scaler = scaler.fit(X_train)

    #Scaling the data
    X_train_scaled = X_scaler.transform(X_train)
    X_test_scaled = X_scaler.transform(X_test)
    
    #Create an XG boost model
    for learning_rate in learning_rates:
        classifier = GradientBoostingClassifier(n_estimators=20,
        learning_rate=learning_rate,
        max_features=5,
        max_depth=3,
        random_state=0)
        classifier.fit(X_train_scaled, y_train)
        acc_score_booster = classifier.score(X_test_scaled, y_test)
    #Create a random forest classifier
    rf_model = RandomForestClassifier(n_estimators=100, random_state=5)
    
    # Fitting the model
    rf_model = rf_model.fit(X_train_scaled, y_train)
    
    
    #Making predictions using testing data
    predictions = rf_model.predict(X_test_scaled)
    
    #Calculating the confusion matrix
    cm = confusion_matrix(y_test, predictions)

    cm_df = pd.DataFrame(cm, index=["Actual 0(decrease)", "Actual 1(increase)"], 
                         columns = ["Predicted 0(decrease)", "Predicted 1(increase)"])
    
    #calculate accuracy
    acc_score = accuracy_score(y_test, predictions)
    #Print confusion matrix for each stock
    print(f"{companies[i]:}\n"
          f"-----------\n"
          f"{cm_df}\n"
          f"Accuracy: {acc_score}\n"
          f"Learning rate: {learning_rate}\n"
         f"Boosting Accuracy: {acc_score_booster}")
    acc_dict.append(acc_score)
    boost_dict.append(acc_score_booster)
```

```python
fig_df = pd.DataFrame({"companies": companies, "accuracy score":acc_dict, "boosted score": acc_score_booster})
fig_df.sort_values(by="accuracy score",ascending=False)
```
